# Load, display and save images

## Load image from file

```python
import cv2
import matplotlib.pyplot as plt

b, g, r = cv2.split(cv2.imread('wave.png'))  # use 0 for greyscale image and 1 for colored image
image = cv2.merge([r, g, b])
type(image)  # image is stored as numpy array
```

## Display image

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

## Save image to file

```python
cv2.imwrite('wave.jpg', image)  # convert image to jpg and save
```
